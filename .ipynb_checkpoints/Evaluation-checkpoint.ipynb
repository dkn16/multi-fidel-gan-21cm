```python
import numpy as np
device = "cuda"
n_latent = 512
n_mlp = 8

#this file contains the corrsponding redshift of each slice in our dataset
redshifts = np.load('/scratch/dkn16/redshifts_256.npy')

g_ema = Generator(
    64, 512, 8, channel_multiplier=2,extend=True
).to(device)

#here we load our model. the model can be shared upon request to the author's email
checkpoint = torch.load('../few-shot-gan-tbandrho/checkpoints/lc_64_to_256_fp/002800.pt')

#for data parallel
g_ema = torch.nn.DataParallel(g_ema)

g_ema.load_state_dict(checkpoint["g_ema"],strict=False)
```

```python
import random
import numpy as np
import re

#set the director to our test set
direc='/scratch/dkn16/dataset_lc512/data-tbandrho/'
#set a re formula to get the model parameters from the filename 
pattern = re.compile(r'\d+.\d+')
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D, HandlerTuple

index1=1
params=[]
global_test=[]
global_gan=[]
filenamesbase = ['/scratch/dkn16/lightcone_test/tbandrho_256/lightcone_22.3614_5.5_256_',
                '/scratch/dkn16/lightcone_test/tbandrho_256/lightcone_50.03_5.0_256_',
                '/scratch/dkn16/lightcone_test/tbandrho_256/lightcone_111.815_4.5_256_']
index1=[17,5,13]
total_test=[]
total_rho = []
total_GAN = []
total_GANrho = []


for i in range(3):
    print(i)
    
    #here we load our test set
    # get the filenames in our dataset 
    for index in range(4):
        filename = filenamesbase[i]+str(int(index1[i]+index))+'.npy'
        datacache=np.load(filename)
        if index==0:
            testdata = datacache[0]
            testdata=np.concatenate((testdata[::2],testdata.transpose(1,0,2)),axis=0)
            testrho = datacache[1]
            testrho=np.concatenate((testrho,testrho.transpose(1,0,2)[1::2]),axis=0)
            print(filename)
        else:
            testdata=np.concatenate((testdata,datacache[0][::2]),axis=0)
            testdata=np.concatenate((testdata,datacache[0].transpose(1,0,2)[1::2]),axis=0)
            testrho=np.concatenate((testrho,datacache[1][::2]),axis=0)
            testrho=np.concatenate((testrho,datacache[1].transpose(1,0,2)[1::2]),axis=0)
            
    global_test.append(np.mean(testdata,axis=(0,1)))
    total_test.append(testdata)
    total_rho.append(testrho)
    
    result = pattern.findall(filename)
    testparams=np.array([float(result[2]),np.log10(float(result[1]))])
    print(testparams)
    params.append(testparams)
    
    #here we generate the GAN samples
    for j in range(50):
        normalized_param=(testparams-np.array([4,1]))/np.array([2,1.398])
        #normalized_param=[0.4,0.4]
        sample_z = torch.randn(16, 512, device=device)
        sample_z1 = torch.randn(16, 512, device=device)
        repeat_param=np.tile(normalized_param,16).reshape((16,-1))
        repeat_param=torch.tensor(repeat_param,device=device).float()
        with torch.no_grad():
            g_ema.eval()
            GANsample, _ = g_ema(repeat_param,[sample_z,sample_z1])
            GANsample=GANsample.to('cpu').numpy()
        #GANsample[GANsample<0]=0
        if j==0:
            GANdata=GANsample
        else:
            GANdata=np.concatenate((GANdata,GANsample),axis=0)
    #GANdata[GANdata<0]=0
    GANRho = (GANdata[:,1,...]+1.)*2
    GANdata = GANdata[:,0,...]*30
    GANdata[GANdata<0]=0
    
    global_gan.append(np.mean(GANdata,axis=(0,1)))
    total_GAN.append(GANdata)
    total_GANrho.append(GANRho)
```

```python
from kymatio.torch import Scattering2D
from matplotlib.legend_handler import HandlerLine2D, HandlerTuple
redshifts = np.load('/scratch/dkn16/redshifts_256.npy')
nchunks=6
GANtbtorch = torch.tensor(total_GAN[0][:,:,nchunks*64:(nchunks+2)*64],device = "cuda",dtype = torch.float32)


# Set the parameters of the scattering transform.
J = 7
M, N = 256,128
S = Scattering2D(J, (M, N),L=4,out_type='list').cuda()
jthetas = []
for dicts in S(GANtbtorch):
    jthetas.append([dicts['j'],dicts['theta']])
jthetas = np.array(jthetas)
xlables = []
# Define a Scattering2D object.

S = Scattering2D(J, (M, N),L=4).cuda()

Sgan = []
Stest = []
for i in range(3):
    GANtbtorch = torch.tensor(total_GAN[i][:,:,nchunks*64:(nchunks+2)*64],device = "cuda",dtype = torch.float32)
    Sx1 = np.mean(S(GANtbtorch).to('cpu').numpy(),axis=(2,3))
    for j in [0,3,6]:
        index = [a[0] == (j,) for a in jthetas]
        #print(index)
        cache = Sx1[:,index]
        #print(cache.shape)
        if j == 0:
            S1 = cache
        else:
            S1=np.concatenate((S1,cache),axis = 1)
        #xlables.append((j))
    for j in [0,3,6]:
        for j2 in [0,3,6]:
            if j2<j:
                continue
            index = [a[0] == (j,j2) for a in jthetas]
            cache = Sx1[:,index]
            S1=np.concatenate((S1,cache),axis = 1)
            #xlables.append((j2,j))
    print(S1.shape)

    testtorch = torch.tensor(total_test[i][:,:,nchunks*64:(nchunks+2)*64],device = "cuda",dtype = torch.float32)
    Sx1 = np.mean(S(testtorch).to('cpu').numpy(),axis=(2,3))

    for j in [0,3,6]:
        index = [a[0] == (j,) for a in jthetas]
        #print(index)
        cache = Sx1[:,index]
        #print(cache.shape)
        if j == 0:
            St = cache
        else:
            St=np.concatenate((St,cache),axis = 1)
        #xlables.append((j))
    for j in [0,3,6]:
        for j2 in [0,3,6]:
            if j2<j:
                continue
            index = [a[0] == (j,j2) for a in jthetas]
            cache = Sx1[:,index]
            St=np.concatenate((St,cache),axis = 1)
            #xlables.append((j2,j))
    St.shape
    Sgan.append(S1)
    Stest.append(St)
```
